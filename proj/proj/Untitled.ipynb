```python
import pandas as pd
```

```python
hotel_df=pd.read_csv('realdata.csv')
hotel_df
```

```python
hotelnamelocdf=hotel_df[["HOTEL NAME","LOCATION"]]
hotelnamelocdf
```

```python
mainhotelnamelocdf=hotelnamelocdf.drop_duplicates()
mainhotelnamelocdf
```

```python
hotelrating=hotel_df.pivot_table(index=['ID'],columns=['HOTEL NAME'],values='RATING')
hotelrating
```

```python
corrMatrix=hotelrating.corr(method='pearson')
corrMatrix
```

```python
userrating=hotelrating.loc[1].dropna()
userrating
```

```python
simcandidates = pd.Series()
#simcandidates=pd.DataFrame()
for i in range(0, len(userrating.index)):
    print(userrating.index[i]) 
    #retrive similar hotel to this one i rated
    sims = corrMatrix[userrating.index[i]].dropna()
    #scale its similarity by how well i rated this hotel
    sims = sims.map(lambda x: x * userrating[i])
    #add the score to the list of similarity candidates
    simcandidates = simcandidates.append(sims)
    

#glance at our results so far:
simcandidates
```

```python
print( "sorting...")
simcandidates.sort_values(inplace = True, ascending = False)
simcandidates
```

```python
simcandidates=simcandidates.groupby(simcandidates.index).sum()
simcandidates
```

```python
simcandidates.sort_values(inplace = True, ascending = False)
simcandidates
```

```python
filteredSims=simcandidates.drop(userrating.index)
filteredSims
```

```python
simcandidatesdf=filteredSims.to_frame(name= 'hotels')
simcandidatesdf
```

```python

simcandidatesdf=pd.DataFrame({'HOTEL NAME':filteredSims.index,'rating':filteredSims.values})
simcandidatesdf
```

```python
#newdf=mainhotelnamelocdf.merge
mergedsimdf=pd.merge(simcandidatesdf,mainhotelnamelocdf)
mergedsimdf
```
